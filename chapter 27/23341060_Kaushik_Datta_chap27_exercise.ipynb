```python
import simpy
import random
```

```python
class CarEngine:
    def __init__(self, env):
        self.env = env
        self.temperature = 20  # Initial temperature in degrees Celsius
        self.thermostat = env.process(self.thermostat_control())

    def thermostat_control(self):
        while True:
            yield self.env.timeout(1)  # Check temperature every second

            # Simple thermostat control logic
            if self.temperature > 90:
                print(f"{self.env.now:.2f} - Engine temperature is too high! Cooling down.")
                yield self.env.timeout(random.uniform(2, 5))  # Simulate cooling time
                print(f"{self.env.now:.2f} - Engine temperature is now {self.temperature:.2f} degrees.")
            elif self.temperature < 70:
                print(f"{self.env.now:.2f} - Engine temperature is too low! Warming up.")
                yield self.env.timeout(random.uniform(2, 5))  # Simulate heating time
                print(f"{self.env.now:.2f} - Engine temperature is now {self.temperature:.2f} degrees.")

def car_simulation(env, run_time):
    car = CarEngine(env)
    yield env.timeout(run_time)

def main():
    random.seed(42)
    env = simpy.Environment()
    simulation_time = 20  # Total simulation time in seconds

    print("Car Engine Simulation\n")
    env.process(car_simulation(env, simulation_time))
    env.run()

if __name__ == "__main__":
    main()
```
